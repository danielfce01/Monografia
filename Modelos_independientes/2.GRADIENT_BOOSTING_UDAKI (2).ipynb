###Gradient Boosting

####GB df_aki

```python
#Variables predictoras y objetivo
X_aki = df_aki.drop("lra", axis=1)
y_aki = df_aki["lra"]

#Dividir en entrenamiento y prueba
X_train_aki, X_test_aki, y_train_aki, y_test_aki = train_test_split(
    X_aki, y_aki, test_size=0.3, random_state=42, stratify=y_aki
)


#Escalamiento (normalización)
scaler = StandardScaler()
X_train_scaled_aki = scaler.fit_transform(X_train_aki)
X_test_scaled_aki = scaler.transform(X_test_aki)
```

```python
# Modelo Gradient Boosting
gb = GradientBoostingClassifier(
    n_estimators=200,      # número de árboles
    learning_rate=0.05,    # velocidad de aprendizaje
    max_depth=3,           # profundidad de los árboles base
    random_state=42
)

gb.fit(X_train_scaled_aki, y_train_aki)

# Predicciones
y_pred = gb.predict(X_test_scaled_aki)
y_prob = gb.predict_proba(X_test_scaled_aki)[:, 1]

# Evaluación
print("MATRIZ DE CONFUSIÓN:")
print(confusion_matrix(y_test_aki, y_pred))
print("\n REPORTE DE CLASIFICACIÓN:")
print(classification_report(y_test_aki, y_pred))
print(" ROC-AUC:", roc_auc_score(y_test_aki, y_prob))

# Curva ROC
RocCurveDisplay.from_estimator(gb, X_test_scaled_aki, y_test_aki)
plt.title("Curva ROC - Gradient Boosting")
plt.show()

# Importancia de las variables
importances = gb.feature_importances_
features = X_aki.columns
imp_df = pd.DataFrame({"Variable": features, "Importancia": importances})
imp_df = imp_df.sort_values(by="Importancia", ascending=False)

print("\n Importancia de las variables:")
print(imp_df)

# Visualización
plt.figure(figsize=(8,5))
sns.barplot(data=imp_df, x="Importancia", y="Variable", palette="viridis")
plt.title("Importancia de Variables - Gradient Boosting")
plt.show()
```

**df_UDAKI  con hiperparametros**

```python
# Librerías necesarias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, roc_auc_score

# -----------------------------
# 1. Definir variables predictoras y target
# -----------------------------
X = df_udaki.drop("lra", axis=1)
y = df_udaki["lra"]

# -----------------------------
# 2. Dividir en entrenamiento y prueba
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# -----------------------------
# 3. Escalamiento
# -----------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# -----------------------------
# 4. Definir el modelo base
# -----------------------------
gb = GradientBoostingClassifier(random_state=42)

# -----------------------------
# 5. Definir la grilla de hiperparámetros
# -----------------------------
param_grid = {
    "n_estimators": [100, 200, 300],       # número de árboles
    "learning_rate": [0.01, 0.05, 0.1],    # tasa de aprendizaje
    "max_depth": [2, 3, 4],                # profundidad máxima de cada árbol
    "subsample": [0.8, 1.0],               # fracción de muestras usadas en cada árbol
    "min_samples_split": [2, 5, 10]        # mínimo de muestras para dividir un nodo
}

# -----------------------------
# 6. Configurar GridSearchCV
# -----------------------------
grid_search = GridSearchCV(
    estimator=gb,
    param_grid=param_grid,
    scoring="roc_auc",   # métrica clínica relevante
    cv=5,                # validación cruzada en 5 pliegues
    n_jobs=-1,           # usar todos los núcleos disponibles
    verbose=2
)

# -----------------------------
# 7. Entrenar búsqueda
# -----------------------------
grid_search.fit(X_train_scaled, y_train)

# -----------------------------
# 8. Resultados
# -----------------------------
print("Mejores hiperparámetros encontrados:")
print(grid_search.best_params_)

print("\nMejor ROC-AUC en validación cruzada:")
print(grid_search.best_score_)

# -----------------------------
# 9. Evaluación en conjunto de prueba
# -----------------------------
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_scaled)
y_prob = best_model.predict_proba(X_test_scaled)[:, 1]

print("\nReporte de clasificación en test:")
print(classification_report(y_test, y_pred))
print("ROC-AUC en test:", roc_auc_score(y_test, y_prob))
```

**Analisis**

**Hiperparametros encontrados**

learning_rate = 0.01

→ Muy bajo, lo que significa que cada árbol contribuye poco al modelo final. Esto obliga a usar más árboles, pero mejora la generalización y reduce el riesgo de sobreajuste.

max_depth = 3

→ Árboles poco profundos, lo que favorece interpretabilidad y evita reglas demasiado específicas.
→ Se mantiene la simplicidad en las interacciones entre variables (edad, creatinina, ASA, etc.).

min_samples_split = 10

→ Un nodo necesita al menos 10 muestras para dividirse. Esto suaviza el modelo y evita ramas basadas en muy pocos pacientes.
→ En datasets clínicos, este parámetro ayuda a que las reglas tengan soporte estadístico suficiente.

n_estimators = 200

→ Número de árboles intermedio. Con el learning_rate bajo, se necesita un número suficiente de iteraciones para capturar patrones.

→ Balance entre capacidad predictiva y tiempo de entrenamiento.

• subsample = 0.8

→ Cada árbol se entrena con el 80% de las muestras seleccionadas aleatoriamente.
→ Introduce variabilidad y reduce sobreajuste, similar a un bagging parcial.

**Hiperparametros con optimizacion del umbral**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, precision_recall_curve

# -----------------------------
# 1. Definir variables predictoras y target
# -----------------------------
X = df_udaki.drop("lra", axis=1)
y = df_udaki["lra"]

# -----------------------------
# 2. Dividir en entrenamiento y prueba
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# -----------------------------
# 3. Escalamiento
# -----------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# -----------------------------
# 4. Modelo con hiperparámetros óptimos
# -----------------------------
gb_best = GradientBoostingClassifier(
    learning_rate=0.01,
    max_depth=3,
    min_samples_split=10,
    n_estimators=200,
    subsample=0.8,
    random_state=42
)

gb_best.fit(X_train_scaled, y_train)

# -----------------------------
# 5. Predicciones de probabilidad
# -----------------------------
y_prob = gb_best.predict_proba(X_test_scaled)[:, 1]

# -----------------------------
# 6. Optimización del umbral
# -----------------------------
precisions, recalls, thresholds = precision_recall_curve(y_test, y_prob)

# Elegir el umbral que maximiza el recall sin perder demasiado en precisión
# Aquí como ejemplo: el umbral que maximiza F1
f1_scores = 2 * (precisions * recalls) / (precisions + recalls)
best_index = np.argmax(f1_scores)
best_threshold = thresholds[best_index]

print("Umbral óptimo encontrado:", best_threshold)

# -----------------------------
# 7. Predicciones con el nuevo umbral
# -----------------------------
y_pred_opt = (y_prob >= best_threshold).astype(int)

print("\nMatriz de confusión con umbral optimizado:")
print(confusion_matrix(y_test, y_pred_opt))

print("\nReporte de clasificación con umbral optimizado:")
print(classification_report(y_test, y_pred_opt))

print("ROC-AUC en test:", roc_auc_score(y_test, y_prob))


```

**Analisis**

**Matrix de confusion**

Clase 0 (sin LRA): 115 verdaderos negativos, pero 151 falsos positivos.

Clase 1 (con LRA): 179 verdaderos positivos, solo 23 falsos negativos.

→ El modelo ahora detecta la mayoría de los pacientes con LRA, aunque aumenta el número de falsos positivos.

**Reporte de clasificación:**

• 	Clase 0 (sin LRA)

• 	Precision: 0.83

• 	Recall: 0.43

• 	F1: 0.57

→ Se pierde sensibilidad en la clase negativa, porque muchos pacientes sanos son clasificados como LRA.

• 	Clase 1 (con LRA)

• 	Precision: 0.54

• 	Recall: 0.89

• 	F1: 0.67

→ Gran mejora en recall: ahora el modelo captura casi 9 de cada 10 pacientes con LRA, aunque con menor precisión.

**ROC-AUC en test: 0.746**

→ Se mantiene prácticamente igual al valor anterior, lo que confirma que el cambio de umbral afecta la clasificación binaria pero no la capacidad discriminativa global del modelo.
